```python
import os
import pickle

folder = 'pickle'
inner_folder = 'model'

with open(os.path.join(folder, inner_folder, 'model_pickle'), 'rb') as f:
    model = pickle.load(f)
    
with open(os.path.join(folder, inner_folder, 'history_pickle'), 'rb') as f:
    loaded_history = pickle.load(f)
```

Load Data

```python
import matplotlib.pyplot as plt

fig = plt.figure()
plt.plot(loaded_history['loss'], color='teal', label='loss')
plt.plot(loaded_history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(loaded_history['accuracy'], color='teal', label='accuracy')
plt.plot(loaded_history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
with open(os.path.join(folder, inner_folder, 'confusion_data'), 'rb') as f:
    loaded_confusion_data = pickle.load(f)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

loaded_conf_matrix = loaded_confusion_data['conf_matrix']

# Access the class labels
class_names = loaded_confusion_data['class_labels']

plt.figure(figsize=(12, 9))
sns.heatmap(loaded_conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
with open(os.path.join(folder, inner_folder,'normalised_confusion_data'), 'rb') as f:
    loaded_confusion_data = pickle.load(f)

loaded_normalised_conf_matrix = loaded_confusion_data['normalised_conf_matrix']
    
plt.figure(figsize=(12, 9))
sns.heatmap(loaded_normalised_conf_matrix, annot=True, fmt='.2f', cmap='Reds', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Normalised Confusion Matrix')
plt.show()
```
