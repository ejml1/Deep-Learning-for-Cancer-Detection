```python
image_path = 'BM_cytomorphology_data/LYI/LYI_00021.jpg'
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(250, 250))
image_array = tf.keras.preprocessing.image.img_to_array(image)
image_array = np.expand_dims(image_array, axis=0)
```

```python
plt.imshow(image)
plt.axis('off')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
data_gen = ImageDataGenerator(
    rotation_range=180,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.25,
    height_shift_range=0.25,
    shear_range=0.2,
    fill_mode='reflect',
)
```

```python
samples = data_gen.flow(image_array, batch_size=1)

# Plot the original and generated images
plt.figure(figsize=(12, 12))
for i in range(9):
    plt.subplot(3, 3, i+1)
    batch = samples.next()
    image_generated = batch[0].astype('uint8')
    plt.imshow(image_generated)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
